```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
frame = pd.read_csv("digital_currency.csv", index_col=0, parse_dates=True)
```

```python
frame.head()
```

# Побудувати графік зміни ціни на час відкриття біржі (в доларах):
### а) загальний;
### б) за 2019 рік;
### в) за грудень 2020 року;
### г) за жовтень 2018 – липень 2020;
### д) за 2019 та 2020 на одному графіку.

```python
first_frame = pd.Series(frame["open_USD"])
```

```python
first_frame.plot()
plt.title("а) Загальний графік зміни ціни на час відкриття біржі;")
```

```python
first_frame.loc['2019'].plot()
plt.title("б) Графік зміни ціни на час відкриття біржі за 2019 рік;")
```

```python
first_frame.loc['2020-12'].plot()
plt.title("в) Графік зміни ціни на час відкриття біржі за грудень 2020 року;")
```

```python
first_frame.loc['2018-10':"2020-07"].plot()
plt.title("г) Графік зміни ціни на час відкриття біржі за жовтень 2018 - липень 2020 років;")
```

```python
first_frame.loc['2019'].plot(color="PaleGreen")
first_frame.loc['2020'].plot()
plt.title("д) Графік зміни ціни на час відкриття біржі за 2019 - 2020 роки;")
```

# Знайти середні значення найменшої ціни за день (в доларах):
### а) за 2019 рік;
### б) за кожний місяць;
### в) за кожний тиждень весни 2020 року.
### г) Розрахувати і зобразити зміни найменшої ціни за день у відсотках за кожен день впродовж весни 2020 року.
### д) Знайти та зобразити графічно ковзне середнє найменшої ціни за день за 2019 рік з вікном в місяць.


```python
second_frame = pd.Series(frame['low_USD'])
```

```python
second_frame.loc['2019'].mean()  #a
```

```python
second_frame.resample('M').mean()  #б
```

```python
second_frame.loc['2020-03':'2020-06'].resample("W").mean()  #в
```

```python
second_frame.loc['2020-03':'2020-06'].pct_change()  #г
```

```python
second_frame.loc['2020-03':'2020-06'].pct_change().plot()
```

```python
second_frame.loc['2019'].rolling('30D').mean()  #д
```

```python
second_frame.loc['2019'].rolling('30D').mean().plot()
```
